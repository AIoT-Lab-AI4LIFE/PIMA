```python
from transformers import AutoTokenizer, AutoModel
import torch


#Mean Pooling - Take attention mask into account for correct averaging
def mean_pooling(model_output, attention_mask):
    token_embeddings = model_output[0] #First element of model_output contains all token embeddings
    input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
    return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)


# Sentences we want sentence embeddings for
sentences = ['TOA THUỐC BHYT', 'Năm sinh1949', 'Họ và tên ĐỖ VĂN HAY', 'Địa chỉ', 'Nam', '208', 'Nữ', 'Chẩn đoán: I10 - Tăng huyết áp vô căn (nguyên phát); (G45) Cơn thiếu máu não thoáng qua và hội', 'C', 'Xã Công Lý - Lý Nhân - Hà Nam', 'Thẻ bảo hiểm y tế', 'HT', 'Mạch:', '2', 'lần/phút', '35', '35', 'Huyết áp:', '/', '17663', 'mmHg', 'Thân nhiệt:', 'Cận lâm sàng:', 'chứng liên quan', '1) NIFEDIPIN T20 STADA RETARD 20 mg', 'SL:32 Viên', 'Sáng 1 Viên', 'SL:20 Viên', 'Uống:', '2) APRATAM 400mg', 'Sáng 1 Viên', 'Ghi chú', 'Ghi chú', 'Ngày 10/10/2019', 'Uống:', 'Cộng khoản: 2 khoản', 'Chiều 1 Viên', '', 'Ngày hẹn tái khám: Thứ hai, 11/11/2019', 'Lời dặn đơn thuốc:', 'Bệnh nhân ký nhận', 'Y, Bác sĩ điều trị']

# Load model from HuggingFace Hub
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/paraphrase-mpnet-base-v2')
model = AutoModel.from_pretrained('sentence-transformers/paraphrase-mpnet-base-v2')

# Tokenize sentences
encoded_input = tokenizer(sentences, max_length=32, padding='max_length', truncation=True, return_tensors='pt')

# Compute token embeddings
model_output = model(encoded_input['input_ids'], encoded_input['attention_mask'])

# Perform pooling. In this case, max pooling.
sentence_embeddings = mean_pooling(model_output, encoded_input['attention_mask'])

print("Sentence embeddings:")
print(sentence_embeddings)
```

```python
class sentencesTransformer(torch.nn.Module):
    def __init__(self, model_name='sentence-transformers/paraphrase-mpnet-base-v2', trainable=False):
        super().__init__()

        self.model = AutoModel.from_pretrained(model_name)
        for param in self.model.parameters():
            param.requires_grad = trainable

    def mean_pooling(self, model_output, attention_mask):
        token_embeddings = model_output[0] #First element of model_output contains all token embeddings
        input_mask_expanded = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
        return torch.sum(token_embeddings * input_mask_expanded, 1) / torch.clamp(input_mask_expanded.sum(1), min=1e-9)

    def forward(self, text_sentences_ids, text_sentences_mask):
        model_output = self.model(text_sentences_ids, text_sentences_mask)
        return self.mean_pooling(model_output, text_sentences_mask)
```
